### DATA CLEANING CASE 1

```python
#Definimos subset
data1=dataset[["Country", "People and Society: Infant mortality rate - total", "People and Society: Tobacco use - female", "People and Society: Children under the age of 5 years underweight", "People and Society: Obesity - adult prevalence rate"]]
#Cambio nombre columnas
dataset1=data1.rename({"People and Society: Infant mortality rate - total":"Infant mortality rate", "People and Society: Tobacco use - female":"Tobacco use female", "People and Society: Children under the age of 5 years underweight":"Children underweight", "People and Society: Obesity - adult prevalence rate":"Obesity adult rate"}, axis=1)
#Borramos NaN
dataset1=dataset1.dropna()
dataset1.head()
```

```python
dataset1
```

```python
#De la primera columna quedarse solo hasta que pone deaths y/o convertirlo a porcentaje, del resto el porcentaje
j=0
for i in dataset1.iloc[:, 1]:
    pos=i.find("deaths")
    num=float(i[:pos-1])
    res=num/10
    dataset1.iloc[j,1]=res
    j+=1

j=0
for i in dataset1.iloc[:, 2]:
    pos=i.find("%")
    num=float(i[:pos])
    res=num/10
    dataset1.iloc[j,2]=res
    j+=1
    
j=0
for i in dataset1.iloc[:, 3]:
    pos=i.find("%")
    num=float(i[:pos])
    res=num/10
    dataset1.iloc[j,3]=res
    j+=1

j=0
for i in dataset1.iloc[:, 4]:
    pos=i.find("%")
    num=float(i[:pos])
    res=num/10
    dataset1.iloc[j,4]=res
    j+=1
```

```python
dataset1.head()
```

```python
dataset1.to_csv(r'case1.csv')
```

### DATA CLEANING CASE 2

```python
#Definimos subset
data2=dataset[["Country", "People and Society: Life expectancy at birth - total population", "Environment: Air pollutants - particulate matter emissions", "Environment: Air pollutants - carbon dioxide emissions", "Environment: Air pollutants - methane emissions"]]
#Cambio nombre columnas
dataset2=data2.rename({"People and Society: Life expectancy at birth - total population":"Life expectancy", "Environment: Air pollutants - particulate matter emissions":"particulate matter emissions", "Environment: Air pollutants - carbon dioxide emissions":"carbon dioxide emissions", "Environment: Air pollutants - methane emissions":"methane emissions"}, axis=1)
#Borramos NaN
dataset2=dataset2.dropna()
```

```python
j=0
for i in dataset2.iloc[:, 1]:
    pos=i.find("years")
    num=float(i[:pos-1])
    dataset2.iloc[j,1]=num
    j+=1

j=0
for i in dataset2.iloc[:, 2]:
    pos=i.find("micrograms")
    num=float(i[:pos-1])
    res=num/10
    dataset2.iloc[j,2]=res
    j+=1
    
j=0
for i in dataset2.iloc[:, 3]:
    pos=i.find("megatons")
    num=i[:pos-1]
    if "," in num:
        num=float(num.replace(",",""))
    else:
        num=float(num)
    res=num/10
    dataset2.iloc[j,3]=res
    j+=1

j=0
for i in dataset2.iloc[:, 4]:
    pos=i.find("megatons")
    num=i[:pos-1]
    if "," in num:
        num=float(num.replace(",",""))
    else:
        num=float(num)
    res=num/10
    dataset2.iloc[j,4]=res
    j+=1
```

```python
dataset2.head()
```

```python
data2=dataset2[["Life expectancy","particulate matter emissions","carbon dioxide emissions","methane emissions"]]
corr = data2.corr()
corr
```

```python
dataset2.to_csv(r'case2.csv')
```

### DATA CLEANING CASE 4

```python
#Definimos subset
data4=dataset[["Country", "Economy: Unemployment rate", "People and Society: Education expenditures", "Economy: Inflation rate (consumer prices)", "Economy: Real GDP (purchasing power parity)"]]
#Cambio nombre columnas
dataset4=data4.rename({"Economy: Unemployment rate":"Unemployment rate", "People and Society: Education expenditures":"Education expenditures", "Economy: Inflation rate (consumer prices)": "Inflation rate" ,"Economy: Real GDP (purchasing power parity)":"GDP"}, axis=1)
#Borramos NaN
dataset4=dataset4.dropna()
```

```python
dataset4.head()
```

```python
j=0
for i in dataset4.iloc[:, 1]:
    pos=i.find("%")
    num=float(i[:pos])
    dataset4.iloc[j,1]=num
    j+=1

j=0
for i in dataset4.iloc[:, 2]:
    pos=i.find("%")
    num=float(i[:pos])
    dataset4.iloc[j,2]=num
    j+=1
  
j=0
for i in dataset4.iloc[:, 3]:
    pos=i.find("%")
    num=i[:pos]
    num=num.replace(",","")
    dataset4.iloc[j,3]=float(num)
    j+=1

j=0
for i in dataset4.iloc[:, 4]:
    borrado=0
    pos=i.find("(")
    i2=i[1:pos-1]
    pos=i2.find("b")
    num=i[1:pos-1]
    if pos == (-1):
        borrado=1
        pos=i2.find("m")
        num=i[1:pos-1]
        if pos == (-1):
            borrado=2
            pos=i.find("(")
            num=i[1:pos-1]
            num=num.replace(",","")
            if pos == (-1):
                borrado=4
    if borrado==0:
        dataset4.iloc[j,4]=float(num)
    elif borrado==1:
        dataset4.iloc[j,4]=float(num)/(10**6)
        print(float(num)/(10**6))
    elif borrado==2:
        dataset4.iloc[j,4]=float(num)/(10**12)
        print(float(num)/(10**12))
    else:
        dataset4.iloc[j,4]=None
    j+=1

dataset4=dataset4.dropna()
```
